### Conocemos los datos y ahora qué hacer para prepararlos (limpiarlos)?

Pandas tiene todas las instrucciones para:

Al observar las observaciones no-nulas para cada columna, vemos que a la variable *embarked* (una de nuestras variables categóricas), le faltan únicamente dos datos, por lo que se procederá a calcular la moda y rellenar los valores faltantes.

```python
# Calcular la moda de la columna 'embarked'
moda_embarked = datos['embarked'].mode()[0]

# Llenar los valores faltantes con la moda
datos['embarked'] = datos['embarked'].fillna(moda_embarked)
```

###  Eliminar columnas

La columna "body" normalmente no nos será útil, eliminémosla

```python
newdf = datos.drop(columns= ['body'])
```

```python
newdf.head()
```

### Cambiar el tipo de algunos datos

Varias de las columnas son categóricas, habrá que codificarlas

De acuerdo con los datos, tenemos 3 variables categóricas que pueden necesitar codificación: *survived, sex y embarked* , por lo que hay que codificarlas. Sin embargo, la variable survived ya se encuentra codificada, por lo que solo tenemos que codificar las últimas dos.

```python
#Codificamos la variable 'sex'

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
newdf['sex'] = label_encoder.fit_transform(newdf['sex'])
```

```python
#Codificamos la variable 'embarked'

from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False) 
encoded_embarked = encoder.fit_transform(newdf[['embarked']])
encoded_embarked_df = pd.DataFrame(encoded_embarked, columns=encoder.get_feature_names_out(['embarked']))
newdf = pd.concat([newdf.drop(columns=['embarked']), encoded_embarked_df], axis=1)
```

```python
newdf.head(20)
```

### Estandarizar los nombres de las columnas:

Estandarizamos el nombre de las columnas que codificamos

```python
# Estandarizar los nombres de todas las columnas, utilizando una expresión regular como nuestro estándar
newdf.columns = newdf.columns.str.lower().str.replace(' ', '_').str.replace('[^a-z0-9_]', '', regex=True)

newdf.head()
```

### Reemplazar valores faltantes

```python
newdf.info()
```

Para esto, observamos que a nuestras variables *cabin*,*boat* y *home.dest* les faltan muchos datos. Sin embargo, como no son esenciales para el análisis que queremos realizar, las podemos eliminar también.

```python
datosnew = newdf.drop(columns= ['cabin', 'boat', 'homedest'])
datosnew.head(10)
```

### Reducir la cardinalidad

Pueden haber columas con datos categóricos demasiados variados, se podría poner un límite de frecuencia de una categoría y las que no cumplen cambiarles a un valor ötros", o algo así

Después de analizar las columnas categóricas, se encontró que ninguna de ellas tiene una alta cantidad de valores únicos que justifique la reducción de cardinalidad. Todas las categorías en estas columnas son manejables y no afectan la interpretación de los datos, por lo que no se realizó ninguna modificación en este paso.

### Eliminar datos atípicos **Falta revisar**

```python

# Lista de columnas numéricas a analizar
numeric_cols = ['age', 'fare', 'sibsp', 'parch']

# Detectar y eliminar outliers usando IQR
for col in numeric_cols:
    Q1 = datos[col].quantile(0.25)  
    Q3 = datos[col].quantile(0.75)  
    IQR = Q3 - Q1  

    # Definir los límites para valores atípicos
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filtrar los datos dentro de los límites
    datos = datos[(datos[col] >= lower_bound) & (datos[col] <= upper_bound)]

print("Datos después de eliminar outliers:")
print(datos.shape)
```
